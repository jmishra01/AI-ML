```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Data url & file name
data = "https://raw.githubusercontent.com/jmishra01/AI-ML/refs/heads/main/data/Real-estate.csv"
file_name = "Real-estate.csv"
```

```python
# Dowload csv file using requests library
import requests

res = requests.get(data)
if res.status_code == 200:
    with open(file_name, 'wb') as f:
        f.write(res.content)
```

```python
# Display first 5 rows
!head -n 5 Real-estate.csv
```

```python
# Read csv file, and create pandas dataframe
df = pd.read_csv(file_name)
```

```python
columns = ["No", "trans-date", "house-age", "dist-mrt-stn", "#-stores", "lat", "lng", "price-per-unit"]
df.columns = columns
df = df[columns[1:]]
```

```python
df.head()
```

```python
X = df.drop("price-per-unit", axis=1)
y = df["price-per-unit"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LinearRegression as LR
```

```python
RANDOM_STATE = 42
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=RANDOM_STATE)
```

```python
numeric_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('numeric', numeric_pipe, X.select_dtypes(include=['int64', 'float64']).columns),
    ('categorical', categorical_pipe, X.select_dtypes(include=['object']).columns)
])

full_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LR())
])
```

```python
full_pipe.fit(X_train, y_train)
```

```python
# Predict y_pred values using predict function.
y_pred = full_pipe.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
    )
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
